```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.investing.com/equities/trending-stocks'
```

```python
req = Request(url,headers={'User-Agent':'Mozilla/5.0'})
```

```python
response1 = urlopen(req).read()
```

```python
response1
```

```python
html_soup = BeautifulSoup(response1,"html.parser")
```

```python
type(html_soup)
```

```python
html_soup.select('#trending')
```

```python
company_name = html_soup.find_all('div', class_='chartPopWrap aqlink float_lang_base_1 miniChartChild js-hover-me-wrapper js-hover-me')
```

```python
repr(html_soup.find(id='microChartData').get_text())
```

```python
trend_market =[]
for item in range(0,len(html_soup.find_all(id='microChartData'))):
    trend_market.append(repr(html_soup.find_all(id='microChartData')[item].get_text()).split("\\n"))
```

```python
trend_market
```

```python
import pandas as pd
```

```python
a = pd.DataFrame(trend_market)
```

```python
a.columns=(['c1','c2','c3','c4','c5','c6','c7','c8','c9','c10','c11'])
```

```python
a.drop({'c1','c3','c6','c7','c8','c9','c10','c11'},axis=1)
```

```python
a[['Changes_value','Changes_Percentage']] = pd.DataFrame([a['c5'][i].replace('+',' +').replace('-',' -').split()
                                                          for i in range(0,len(a))])
```
